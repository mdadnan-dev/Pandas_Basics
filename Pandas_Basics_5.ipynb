### Filtering Columns

```python
bios.loc[bios['height_cm'] > 200]

# bios['height_cm']>200  --> this takes the values that are greater than 200
# bios.loc --> finds all the values that are greater than 200 and shows the result
```

```python
bios.loc[bios['weight_kg'] > 90]

# gives values based on locations that are greater than 90
```

```python
bios.loc[bios['height_cm'] > 220 ,['name','height_cm']]

# bios.loc[bios['height_cm'] > 220 --> this gives the values that are greater than 220 
# ['name','height_cm'] --> gives the number of columns to be displayed 
```

```python
bios.loc[bios['height_cm'] > 216][['name','height_cm']]  

# This code works same as the above code, its just an alternative way to write the above the code 
```

```python
bios.loc[(bios['height_cm'] > 210) & (bios['born_country'] == 'CAN'), ['name','height_cm','born_country']]

# ['height_cm'] > 210) & (bios['born_country'] == 'CAN' --> we are adding multiple conditions 
# 1. we are finding values of height that are greater than 210
# 2. we are finding values of born_country that is CAN
# 3. we are using the & [AND] operator to make sure both the conditions are true together [Both the values must be true to be considered in the result]
# ['name','height_cm','born_country']] --> we are printing columns name, height_cm, born_country rather than all the columns in the data
```

```python
bios.loc[(bios['height_cm'] > 210) | (bios['born_country'] == 'USA'), ['name','height_cm','born_country']]

# ['height_cm'] > 210) & (bios['born_country'] == 'USA' --> we are adding multiple conditions 
# 1. we are finding values of height that are greater than 210
# 2. we are finding values of born_country that is USA
# 3. we are using the | (OR) operator to make sure the conditions are true [even if one condition satisfies the result is printed]
# ['name','height_cm','born_country']] --> we are printing columns name, height_cm, born_country rather than all the columns in the data
```

### str.contains

```python
bios.loc[bios['name'].str.contains("Adnan")]  # This gives the values that contains the str Adnan 
```

```python
bios.loc[bios['born_country'].str.contains("US",na = False),['name','born_country']] 

# This gives the str values that contains US in the born_country column
# The na = False does not print the Null values
# The columns name, born_country are shown as output
```

```python
bios.loc[bios['born_country'].str.contains("GE|US",na = False),['name','born_country']]

# This gives the values of str that contains GE or US  [ anyone can be true to execute ]
# na = False --> Null values are not shown in the result 
# name, born_country -> shows these two columns in the result 
```

### isin

```python
bios.loc[bios['born_country'].isin(['USA','RUS','AUS'])]

# here using the isin we are checking if the born_country is present in USA, RUS, AUS
```

### startswith

```python
bios.loc[bios['born_country'].isin(['USA','RUS','AUS']) & bios['name'].str.startswith('Bill')]

# The startswith --> checks if the starting value starts with the specified string name [Bill]
```

```python
bios.loc[bios['name'].str.startswith('A')]

# Here we are accessing the names of the people that starts with the letter A 
```

### query

```python
bios.query('born_country == "USA"')

# the query() gives the user an experience of using SQL environment
```

```python
bios.query('born_country == "IND" and height_cm > 170')
```

```python
bios.query('born_country == "IND" or born_region == "Punjab" ')
```
