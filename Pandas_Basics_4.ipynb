### Updating values 

```python
coffee.head()
```

```python
coffee.loc[1,'Units Sold'] = 40  

# by doing this we can change the value of Units Sold at location 1 
# The original value is 15 
# Now the value is changed to 40
```

```python
coffee.head()
```

```python
coffee.loc[2,'Day'] = 'Friday'

# Here we are assigning the value Friday to location 2 in the Day column
# The original value is Tuesday
# The new value is Friday
```

```python
coffee.head()
```

```python
coffee.loc[1,'Units Sold'] = 15

coffee.head()
```

```python
coffee.loc[2,'Day'] = 'Tuesday'
coffee.head()
```

```python
coffee.loc[0:3,'Units Sold'] = 10


# We can even change values of multiple rows 

coffee.head()
```

```python
coffee.loc[0:3,'Units Sold'] = [25,15,30,20]

coffee.head()
```

# Sorting values 

```python
coffee.sort_values('Day')  # Sorts the values in ascending order based on the Day -> column
```

```python
coffee.sort_values('Units Sold')  # Sorts the values in ascending order based on the Units Sold column
```

```python
coffee.sort_values('Units Sold', ascending=False)  # Gives the values in descending order because we have set the ascending to False
```

```python
coffee.sort_values('Day',ascending= False)  # gives the values in descending order based on the column Day
```

```python
coffee.sort_values(['Day','Coffee Type'])

# gives the ascending values of Day and then ascending values of Coffee Type
```

```python
coffee.sort_values(['Day','Units Sold'], ascending=False)

# gives the values for Day in descending order then gives the values for Units Sold in descending order
```

```python
coffee.sort_values(['Day','Coffee Type'],ascending=[0,1])

# gives descending values for the column Day
# gives the ascending values for the column Coffee Type
```
